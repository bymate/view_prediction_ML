#### for문 돌려서 전체 유튜브 제목 수집

```python
import time
import pandas as pd
from tqdm.notebook import tqdm
from selenium import webdriver
from selenium.webdriver.common.by import By

df = pd.read_csv('youtube_contents_info_final_ver1.csv', index_col=0)

# 제목 리스트화
title_list = []
for t in df['title_text']:
    title_list.append(t)

# 크롬 드라이브 열기
driver = webdriver.Chrome()
driver.get('https://www.youtube.com/')

# 유튜브 타이틀 검색 후 재수집
title_dict = {}
for title in tqdm(title_list):
    # 검색후 입력후 검색
    search = driver.find_element(By.NAME, 'search_query')
    search.clear()
    search.send_keys(title)
    driver.find_element(By.XPATH, '//*[@id="search-icon-legacy"]/yt-icon/span').click()
    time.sleep(1)

    # 첫번째로 발견한 최상단의 제목만 가져옴
    re_title = driver.find_element(By.XPATH, '//*[@id="video-title"]').text

    title_dict[title] = re_title

driver.quit()
```

```python
len(list(title_dict.values()))
```

```python
a=''
for i in list(title_dict.values()):
    if a != i:
        a = i
    else:
        print(i)
```

```python
for idx, i in enumerate(list(title_dict.keys())):
    if '아는누나' in i:
        print(idx, i)
```

```python
df[df['title_text'] == '[sub] 아는누나랑나불 | ??나영석의 나불나불']
```

```python
title_dict['[sub] 아는누나랑나불 | ??나영석의 나불나불'] = '[sub] 아는누나랑나불 | 👅나영석의 나불나불'
```

```python
# 제대로 바꼈는지 확인
for k, v in title_dict.items():
    if '아는누나' in k:
        print(k, '-->', v)
```

```python
df1 = pd.DataFrame(list(title_dict.items()), columns=['o_title', 're_title'])
df1.to_csv('title_re_crawling.csv')
```

```python
driver = webdriver.Chrome()
driver.get('https://www.youtube.com/watch?v=GYnoCCq_4OQ')
```

```python
import time

driver.find_element(By.XPATH, '//*[@id="description-inner"]').click()

# 제목 추출
title = driver.find_element(By.XPATH, '//*[@id="title"]/h1').text

# 화면 스크롤
driver.execute_script('return document.body.scrollHeight')
driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')

# 스크립트 버튼 클릭
button = driver.find_elements(By.XPATH, '//*[@id="primary-button"]/ytd-button-renderer/yt-button-shape')
for i in range(2):
    if button[i].text == '스크립트 표시':
        button[i].click()

time.sleep(1)

# 언어 선택
driver.find_elements(By.XPATH, '//*[@id="label-text"]')[0].click()
time.sleep(1)

# 영어 선택
driver.find_elements(By.XPATH, '//*[@id="menu"]/a[1]/tp-yt-paper-item')[1].click()
time.sleep(1)

script_list_raw = driver.find_elements(By.XPATH, '//*[@id="segments-container"]/ytd-transcript-segment-renderer')
script_list_raw[0].text.split('\n')[1]

script_list = []
for script in script_list_raw:
    s = script.text.split('\n')[1]
    script_list.append(s)

script_list
```

```python
driver.quit()
```

```python
import pandas as pd

dex_an_df = pd.DataFrame({'script': [script_list], 'title_text': [title]})
dex_an_df.to_csv('dex_an_script.csv')
```
