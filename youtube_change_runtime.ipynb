```python
cnt_1 = 0
cnt_2 = 0 
cnt_3 = 0
for t in df['running_time']:
    if len(t.split(':')) == 2:
        cnt_2 += 1
    elif len(t.split(':')) == 3:
        cnt_3 += 1
        if t.split(':')[0] == '2':
            print(t)
    elif len(t.split(':')) == 1:
        cnt_1 += 1
print(cnt_1, cnt_2, cnt_3)
```

```python
# 영상 길이 초 단위로 바꾸기

running_time_list = []
for time in df['running_time']:
    if len(time.split(':')) == 2:
        t = int(time.split(':')[0]) * 60 + int(time.split(':')[1])
        running_time_list.append(t)
    elif len(time.split(':')) == 3:
        if int(time.split(':')[0]) == 1 or int(time.split(':')[0]) == 2:
            t = int(time.split(':')[0]) * 3600 + int(time.split(':')[1]) * 60 + int(time.split(':')[2])
        else:
            t = int(time.split(':')[0]) * 60 + int(time.split(':')[1])
        running_time_list.append(t)

df['running_time(s)'] = running_time_list
df[df['title_text'] == '(live 다시보기) 귀성길은 핑계고']
```

```python
df1 = pd.read_csv('youtube_raw.csv', index_col=0)
df1['running_times(s)'] = running_time_list
max(df1['running_times(s)'])
```
