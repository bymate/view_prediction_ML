```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('guest_info_concat_job_age_ver3.csv', encoding='cp949', index_col=0)
df
```

```python
guest_job_df = pd.read_csv('guest_info_concat.csv', index_col=0)
guest_job_df
```

```python
guest_df = df.merge(guest_job_df[['url', 'job']], on='url', how='inner')
guest_df
```

```python
youtube_raw_df = pd.read_csv('change_runningtime.csv', index_col=0)

# yr_df 중복 url 제거
youtube_raw_df = youtube_raw_df.drop_duplicates(subset=['url'])
youtube_raw_df.head(1)
```

```python
guest_merge_df = guest_df.merge(youtube_raw_df[['url', 'likes', 'views', 'formatted_date', 'comment_count', 'texts', 'running_times(s)', 'channel', 'sub']], on='url', how='inner')
guest_merge_df = guest_merge_df.drop_duplicates(subset=['title_text'])
guest_merge_df.head(2)
```

- 데이터프레임 예쁘게 정리

```python
guest_merge_df.columns
```

```python
#guest_merge_df['job_num'].unique()
```

```python
channel_df_final = guest_merge_df[['channel', 'sub', 'title_text', 'hash_tag', 'likes',
       'views', 'formatted_date', 'comment_count', 'running_times(s)', 'guest', 'sex', 'entertainment', 'actor', 'comedian', 'singer',
       'arts_creator', 'sports', 'media_production', 'business_other', '10', '20', '30', '40', '50', '60', 'texts']]
```

```python
channel_df_final.head(2)
```

- 영어 자막 스크립트 만 필터링

```python
channel_df_final
```

```python
channel_df_final = channel_df_final[channel_df_final['texts'].str.contains(r'[a-zA-Z]', na=False)]
channel_df_final
```

- 컬럼명 몇개 수정

```python
channel_df_final.rename(columns={'sub': 'subscription', 'texts': 'script', 'formatted_date':'upload_date'}, inplace=True)
```

```python
channel_df_final
```

- 날짜 컬럼 날짜 데이터로 변경

```python
channel_df_final['upload_date'] = pd.to_datetime(channel_df_final['upload_date'], errors='coerce')
channel_df_final
```

```python
channel_df_final = channel_df_final.reset_index(drop=True)
channel_df_final
```

```python
channel_df_final.info()
```

```python
channel_df_final.to_csv('youtube_contents_info_final.csv')
```
